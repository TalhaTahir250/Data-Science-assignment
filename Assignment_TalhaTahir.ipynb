## Q1. Arithmetic and Assignment Operators

```python
# Define two variables
a = 10
b = 3

print("=== Arithmetic Operators ===")
print(f"a = {a}, b = {b}")
print(f"Addition       (a + b)  = {a + b}")
print(f"Subtraction    (a - b)  = {a - b}")
print(f"Multiplication (a * b)  = {a * b}")
print(f"Division       (a / b)  = {a / b:.4f}")
print(f"Floor Division (a // b) = {a // b}")
print(f"Modulus        (a % b)  = {a % b}")
print(f"Exponentiation (a ** b) = {a ** b}")

print()
print("=== Assignment Operators ===")
a += 5;  print(f"a += 5  ->  a = {a}")
a -= 3;  print(f"a -= 3  ->  a = {a}")
a *= 2;  print(f"a *= 2  ->  a = {a}")
a /= 4;  print(f"a /= 4  ->  a = {a}")
a //= 2; print(f"a //= 2 ->  a = {a}")
a %= 4;  print(f"a %= 4  ->  a = {a}")
```

## Q2. Comparison and Logical Operators

```python
# Using fixed sample values (replace with input() when running interactively)
x = 7   # change this value as needed
y = 3   # change this value as needed

# To use user input, uncomment the two lines below:
# x = int(input("Enter first integer:  "))
# y = int(input("Enter second integer: "))

print(f"=== Comparison Operators (x={x}, y={y}) ===")
print(f"x == y  ->  {x == y}")
print(f"x != y  ->  {x != y}")
print(f"x >  y  ->  {x >  y}")
print(f"x <  y  ->  {x <  y}")
print(f"x >= y  ->  {x >= y}")
print(f"x <= y  ->  {x <= y}")

print(f"\n=== Logical Operators ===")
print(f"x > 0 and y > 0        ->  {x > 0 and y > 0}")
print(f"x > 0 or  y < 0        ->  {x > 0 or  y < 0}")
print(f"not (x == y)           ->  {not (x == y)}")
print(f"x > 0 and not(x == y) ->  {x > 0 and not (x == y)}")
```

## Q3. Bitwise Operators

```python
# Two integers for bitwise demonstration
p = 12   # binary: 1100
q = 10   # binary: 1010

print(f"p = {p}  ->  binary: {bin(p)}")
print(f"q = {q}  ->  binary: {bin(q)}")
print()
print(f"AND         p & q   = {p & q}   ({bin(p & q)})")
print(f"OR          p | q   = {p | q}   ({bin(p | q)})")
print(f"XOR         p ^ q   = {p ^ q}    ({bin(p ^ q)})")
print(f"NOT         ~p      = {~p}  ({bin(~p)})")
print(f"Left Shift  p << 1  = {p << 1}   ({bin(p << 1)})")
print(f"Right Shift p >> 1  = {p >> 1}    ({bin(p >> 1)})")
```

## Q4. Identity and Membership Operators

```python
# Create a list of numbers
nums = [10, 20, 30, 40, 50]

ref  = nums                   # alias – points to the same object
copy = [10, 20, 30, 40, 50]   # new object with identical values

print("=== Identity Operators ===")
print(f"nums list: {nums}")
print(f"ref  is nums      ->  {ref  is nums}   (same object in memory)")
print(f"copy is nums      ->  {copy is nums}  (different object)")
print(f"copy is not nums  ->  {copy is not nums}")

print(f"\n=== Membership Operators ===")
print(f"30 in nums       ->  {30 in nums}")
print(f"99 in nums       ->  {99 in nums}")
print(f"99 not in nums   ->  {99 not in nums}")
print(f"10 not in nums   ->  {10 not in nums}")
```

## Q5. String Operations

```python
# String variable
sentence = "Data Science is the future of technology"
print(f"Original: {sentence}\n")

# 1. Length of the string
print(f"1) Length             ->  {len(sentence)}")

# 2. Uppercase and lowercase conversion
print(f"2) Uppercase          ->  {sentence.upper()}")
print(f"   Lowercase          ->  {sentence.lower()}")

# 3. Find a specific word using 'in' operator
word = "future"
print(f"3) '{word}' in sentence  ->  {word in sentence}")

# 4. Replace a word using replace()
print(f"4) After replace()    ->  {sentence.replace('future', 'backbone')}")

# 5. Slice a portion of the string
print(f"5) Slice [0:12]       ->  '{sentence[0:12]}'")
print(f"   Slice [16:27]      ->  '{sentence[16:27]}'")

# 6. Reverse the string using slicing
print(f"6) Reversed           ->  '{sentence[::-1]}'")
```
